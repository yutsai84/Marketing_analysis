This is a dendrogram plot process. The dendrogram is used to help visidualize the hierarchical clustering before the number of cluster k in the data is known in advance.

```python
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
%matplotlib inline
np.set_printoptions(precision=5,suppress=True)
```

## Import Data

```python
# generate two clusters: a with 100 points, b with 50:
np.random.seed(4711) # for repeatability of the tutorial
a = np.random.multivariate_normal([10,0],[[3,1],[1,4]],size=[100,])
b = np.random.multivariate_normal([0,20],[[3,1],[1,4]],size=[50,])
X = np.concatenate((a,b),)
print X.shape #150 samples with 2 dimensions
plt.scatter(X[:,0],X[:,1])
plt.show()
```

## Apply the Hierarchical Clustering

```python
# generate the linkage matrix
Z = linkage(X,'ward')
```

```python
from scipy.cluster.hierarchy import cophenet
from scipy.spatial.distance import pdist
c, coph_dists = cophenet(Z,pdist(X))
c
```

## Plot Dendrogram

```python
# calculate full dendrogram
plt.figure(figsize=(25,10))
plt.title("Hierarchical Clustering Dendogram")
plt.xlabel('sample index')
plt.ylabel('distance')
dendrogram(Z, leaf_rotation=90., leaf_font_size=8)
plt.show()
```

- Looking at indicies in the dendrogram shows that the green cluster has indicies greater or equal to 100 while the red one has such smaller than 100.
- The last merge jumps quite a large distance, suggesting that maybe there's just two clusters here.
